```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

# 자동으로 random seed 지정해 줌


# print(np.column_stack(([1,2,3], [4,5,6])))
# column_stack : 전달받은 list를 일렬로 세운 후 나란히 연결
# 출력값:
# [[1 4]
#  [2 5]
#  [3 6]]


#fishes feature
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]


#생선 데이터 list를 2차원 배열로 바꾼 후 무게와 길이를 묶고 정렬
fish_data = np.column_stack((fish_length, fish_weight)) 

#target data 만들기
#다중배열에서 첫 번째 배열을 따라 연결하는 concatenate() 사용하기
fish_target = np.concatenate((np.ones(35), np.zeros(14)))   #bream(1) : 35 / smelt(0) : 14
# print(fish_target) >> [1. 1. 1. 1. .... 0. 0. 0. 0. ...]

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42)
#train_test_split() : 전달되는 list & array 비율에 맞게 train / test set로 구분해 나누어 줌
#train set, test set 무작위로 셔플

# print(train_input.shape, test_input.shape) >> (36, 2) (13, 2)
# print(train_target.shape, test_target.shape) >>(36,) (13,)

# train_data와 test_data를 각각 36개와 13개로 나눔
# print(test_target) >> 13개의 test target
# 일부 데이터 클래스 개수가 적을 시 샘플 골고루 안 섞임

# >> stratify 매개변수에 target 전달 -> class 비율에 맞게 데이터 나눔

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)
# stratify를 이용해 각 data class의 비율에 맞게 골고루 분배

# print(test_target) >> 2.25 : 1 비율로 실제 target의 비율과 근사값 산출 성공


kn.fit(train_input, train_target)
kn.score(test_input, test_target)  # >> 1.0 출력
print(kn.predict([[25, 150]]))  #>> 임의의 input data 예측하여 출력 > [0.] 출력

mean = np.mean(train_input, axis=0) #data average calculate >> axis=0 : 배열의 세로 행 간 평균
std = np.std(train_input, axis=0)  #표준편차 계산  >>axis=1 : 배열의 가로 열 간 평균
print(mean, std)
train_scaled = (train_input - mean) / std   # >> 표준점수 구하는 공식 (원본데이터 - 평균값) / 표준편차
test_scaled = (test_input - mean) / std
# >> broadcasting(브로드캐스팅)


kn.fit(train_scaled, train_target)  # 표준점수로 scale한 훈련 data 다시 train
kn.score(test_scaled, test_target)  #표준점수로 scale한 시험 data 다시 train

new = ([25, 150] - mean) / std

distances, indexes = kn.kneighbors([new])  #train data 중 근접한 이웃 data 산출 index
plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(new[0], new[1], marker='^')    # 임의의 data
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], marker = 'D') #임의의 data의 이웃 data scatter line
# plt.xlim((0, 1000)) #>> x, y 축의 범위(scale)가 매우 다름
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print(distances)

```
